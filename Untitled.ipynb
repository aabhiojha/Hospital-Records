```python
fig = px.histogram(dataframe,
                   x = 'age',
                   marginal = 'box',
                   nbins = 47,
                   title= "Distribution of Age",
                   width=800,
                   height=800
                   )

fig.update_layout(bargap = 0.01)
fig.show()
```

```python
px.scatter(dataframe,
           x='charges',
           y='bmi',
           facet_col='sex',
           color='smoker'
           )
```

```python
smoker_data = dataframe[dataframe['smoker'] == 'yes']
non_smoker_data = dataframe[dataframe['smoker'] == 'no']
smoker_mean_cost = smoker_data['charges'].mean()
non_smoker_mean_cost = non_smoker_data['charges'].mean()
smoker_mean_cost-non_smoker_mean_cost
```

```python
fig = px.histogram(dataframe,
                   x='charges',
                   marginal='box',
                   color='smoker',
                   title='Annual Medical Charges',
                   width=800,
                   height=600
                   )
fig.update_layout(bargap=0.01)
fig.show() 
```
